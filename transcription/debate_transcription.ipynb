```python
def vid_to_audio(url, filename):
    ydl_opts = {
        'format': 'bestaudio/best',
        'outtmpl': f'{filename}.%(ext)s',
        
        'quiet': True,              
        'no_warnings': True,        
        'noprogress': True,
        
        'fragment_retries': 15,          
        'retry_on_http_error': True, 
        'extractor_args': {
            'youtube': {
                'player_client': ['ios', 'android'],
                'skip': ['web_safari']
            }
        },
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'm4a',
            'preferredquality': '192'
        }]
    }
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        error_code = ydl.download([url])
    return error_code
```

```python
def trim_transcript(source_file_name, line_a, line_b, final_file_name=None):  
    if final_file_name is None:
        # in most cases the new txt file will have the same name as the original
        final_file_name = source_file_name
    source_file = source_file_name+'.txt'
    final_file = trim_dest+final_file_name+'.txt'
    ff = None
    with open(source_file) as sf:
        # open source file
        for i, line in enumerate(sf):
            if i == line_a-1:
                # start writing to the final file
                ff = open(final_file, "w")
            elif i == line_b:
                # end writing to the final file
                if ff:
                    ff.close()
                break
            if ff:
                ff.write(line)
        if ff:
            ff.close()
                    
```

```python
#replace using regex
def add_speakers(deb):
    dest_folder = pathlib.Path(edited_dest)
    #obtain input text
    for path in pathlib.Path(trim_dest).glob(deb+'.txt'):
        input_text = path.read_text()
    
        #create participant map
        participant_map = {}
        for i in range(len(participants[deb])):
            participant_map[participant_ids[deb][i]] = participants[deb][i]   
        # track no of questioners and their ids
        questioner_count = 0
        questioner_reg = {}
        #define replacement func
        def replace_tag(match):
            nonlocal questioner_count
            xx = match.group(1)
            if xx in participant_map:
                return participant_map[xx]
            elif xx in questioner_reg:
                return questioner_reg[xx] 
            else:
                new_label = f"QUESTIONER_{questioner_count}"
                questioner_reg[xx] = new_label
                questioner_count+=1
                return new_label

        pattern = r"\[SPEAKER_(\d{2})\]"
        output_text = re.sub(pattern, replace_tag, input_text)
        new_file_path = dest_folder/(deb+'.txt')
        new_file_path.write_text(output_text)
```

```python
# apply to transcripts
for deb in participants:
    add_speakers(deb)
```
