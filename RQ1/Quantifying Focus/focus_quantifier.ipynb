```python
def find_debate(deb, structure, processing, grouping, person=None):  
    '''
    Easily find pre-processed debate transcript text in desired format
    deb - name of debate from options
    structure - 'separate'(each speaker) / 'combined'(all speakers)
    processing - 'proc'(pre-processing applied) / 'raw'(raw content)
    grouping - 'grouped'(all speaker turns combined) / 'turns'(separated into speaker turns)
    person - name of speaker in debate
    '''
    if structure == 'separate':
        #use sep and grouped dicts
        if grouping == 'turns':
            #use sep dicts
            if processing == 'proc':
                if person != None:                    
                    if person in participants[deb]:
                        return sep_proc_debates[deb][person]
                    else:                        
                        print("Invalid 'person' param. Non-participant in debate")
                        return
                else:                    
                    return sep_proc_debates[deb]
            elif processing == 'raw':
                if person != None:
                    if person in participants[deb]:
                        return sep_raw_debates[deb][person]
                    else:
                        print("Invalid 'person' param. Non-participant in debate")
                        return
                else:                    
                    return sep_raw_debates[deb]
            else:
                print("Invalid 'processing' param. Use options given in docstring")
                return
        elif grouping == 'grouped':
            #use grouped dicts
            if processing == 'proc':
                if person != None:
                    if person in participants[deb]:
                        return grouped_sep_proc[deb][person]
                    else:
                        print("Invalid 'person' param. Non-participant in debate")
                        return
                else:                    
                    return grouped_sep_proc[deb]
            elif processing == 'raw':
                if person != None:
                    if person in participants[deb]:
                        return grouped_sep_raw[deb][person]
                    else:
                        print("Invalid 'person' param. Non-participant in debate")
                        return
                else:                    
                    return grouped_sep_raw[deb]
            else:
                print("Invalid 'processing' param. Use options given in docstring")
                return
        else:
            print("Invalid 'grouping' param. Use options given in docstring")
            return
    elif structure == 'combined':
        #use comb dicts
        if processing == 'proc':
            return comb_proc_debates[deb]
        elif processing == 'raw':
            return comb_raw_debates[deb]
        else:
            print("Invalid 'processing' param. Use options given in docstring")
            return
    else:
        print("Invalid 'structure' param. Use options given in docstring")
        return
    
```

```python
def create_debate_dataframe(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        content = f.read()
        
    pattern = r'^(\d+)\n(.*?)(?=\n\d+\n|\Z)'

    matches = re.findall(pattern, content, re.DOTALL | re.MULTILINE)

    df = pd.DataFrame(matches, columns=['id', 'text'])
    
    df['text'] = df['text'].str.strip()
    
    df['id'] = df['id'].astype(int)
    
    return df
```

* Proportion of turns on topic
* Proportion of words on topic
* Average length of turns on topic (mean words per turn)
* Rank of topic by turns
* Rank of topic by words
* Lexical diversity when discussing topic(TTR or MSTTR)

```python
# creating a dataframe of 'topic focus metrics'
topic_focus_metrics =  []

debates = []
candidates = []
all_topics = []

# proportion of turns on topic = no of turns with that topic / total no of turns
prop_turns = []
```

```python
topic_focus_metrics =  []
for deb in deb_options:    
    for candidate in participants[deb]:
        no_of_turns = 0
        turns_per_topic = {}
        
        total_word_count = 0
        words_per_topic = {}
        lengths_per_topic = {}
        
        for turn in find_debate(deb, 'separate', 'raw', 'turns', candidate):   
            no_of_turns+=1
            total_word_count += len(turn.split())
            for topic in ast.literal_eval(topics_df.loc[topics_df['text'] == turn, 'claim_labels'].values[0]):
                if topic in turns_per_topic:
                    turns_per_topic[topic]+=1
                else:
                    turns_per_topic[topic] = 1
                    
                if topic in words_per_topic:
                    words_per_topic[topic] += len(turn.split())
                else:
                    words_per_topic[topic]=len(turn.split())
                
                if topic in lengths_per_topic:
                    lengths_per_topic[topic].append(len(turn.split()))
                else:
                    lengths_per_topic[topic] = [len(turn.split())]
        
        # create rankings of turns per topic
        rank_turns = {}
        sorted_items = sorted(turns_per_topic.items(), key=lambda x: x[1], reverse=True)        
        rank = 1        
        for value, group in groupby(sorted_items, lambda x: x[1]):
            group_list = list(group)
            for key, val in group_list:
                rank_turns[key] = rank            
            rank += len(group_list)
        # create rankings of words per topic
        rank_words = {}
        sorted_items = sorted(words_per_topic.items(), key=lambda x: x[1], reverse=True)        
        rank = 1        
        for value, group in groupby(sorted_items, lambda x: x[1]):
            group_list = list(group)
            for key, val in group_list:
                rank_words[key] = rank            
            rank += len(group_list)
        
        
        # done with candidate, add metrics        
        for t in turns_per_topic:
            new_row = {}
            new_row['debate']=deb
            new_row['participant']=candidate
            new_row['topic']=t
            
            # calculate proportion of turns in which each topic was present
            prop_turns = turns_per_topic[t]/no_of_turns             
            new_row['prop_turns']=prop_turns
            # calculate proportion of words on topic
            prop_words = words_per_topic[t]/total_word_count
            new_row['prop_words']=prop_words
            # calculate average len of turn on topic
            avg_turn_length = np.mean(lengths_per_topic[t])
            new_row['avg_turn_length'] = avg_turn_length
            # get rank of topic by turns
            new_row['turn_rank']=rank_turns[t]
            # get rank of topic by words
            new_row['word_rank']=rank_words[t]
            
            
            topic_focus_metrics.append(new_row)
```
