```python
%%notify
# Testing ILP plain
sys.argv = ['ILP_decoder.py', 'bert_plain_test_output.csv']

from ILP_decoder import main

input_probs = 'bert_plain_test_output.csv'

main(input_probs)
```

```python
#Evaluating ILP plain

df_ilp = pd.read_csv('bert_plain_test_output_ILP.csv')
label_columns = [col for col in df_ilp.columns if col != 'Input']
f1_preds_ilp = df_ilp[label_columns].values
f1_preds_ilp = f1_preds_ilp.astype(int)

df_gold = pd.read_csv('../data/test.csv')
golds = df_gold[label_columns].values
golds = golds.astype(int)

ix2label = {ix:int(x) for ix,x in enumerate(label_columns)}

results = metrics(f1_preds_ilp, golds, ix2label)
print(results)
```
